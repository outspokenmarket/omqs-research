# 3. Condicionais: ensinando o código a decidir

Condicionais permitem que o programa tome decisões.

Essa lógica aparece diretamente em roteadores determinísticos de agentes.

````python
def logical_and(a, b):
    return a and b

def logical_or(a, b):
    return a or b

# Variáveis booleanas
variables = [True, False]

print("### Tabela Verdade para AND (E) ###")
print("| A     | B     | A AND B |")
print("|-------|-------|---------|")
for a in variables:
    for b in variables:
        result = logical_and(a, b)
        print(f"| {str(a):<5} | {str(b):<5} | {str(result):<7} |")

print("\n### Tabela Verdade para OR (OU) ###")
print("| A     | B     | A OR B  |")
print("|-------|-------|---------|")
for a in variables:
    for b in variables:
        result = logical_or(a, b)
        print(f"| {str(a):<5} | {str(b):<5} | {str(result):<7} |")
````

## 5.1 Uma decisão de agente como dicionário

Um agente pode transformar a pergunta do usuário em algo assim:

```python
{
    "tool": "list_distinct_csv",
    "args": {
        "filename": "sales.csv",
        "column": "product"
    }
}
```

Isso significa:

- `tool`: qual função executar
- `args`: quais argumentos passar para a função

````python
decision = {
    "tool": "list_distinct_csv",
    "args": {
        "filename": "sales.csv",
        "column": "product"
    }
}

print(decision)
print(decision["tool"])
print(decision["args"])
print(decision["args"]["column"])
````

## 6.1 Parâmetros e retorno

Uma função pode receber parâmetros e devolver um resultado com `return`.

Vamos criar uma função que lista valores distintos.

````python
def list_distinct_values(values):
    # set(values) remove duplicatas.
    # sorted() ordena os valores.
    unique_values = sorted(set(values))
    return unique_values
````

````python
products = ["Laptop", "Phone", "Tablet", "Phone", "Laptop", "Borrosca"]

result = list_distinct_values(products)

print(result)
````

## 6.2 Função que retorna texto formatado

Em agentes, normalmente queremos devolver uma resposta em texto.

````python
def list_distinct_values_as_text(values):
    unique_values = sorted(set(values))

    # "\n".join(...) junta os itens colocando uma quebra de linha entre eles.
    return "Distinct values:\n" + "\n".join(unique_values)

products = ["Laptop", "Phone", "Tablet", "Phone", "Laptop", "Borrosca"]

print(list_distinct_values_as_text(products))
````

## 6.3 Docstrings

Uma docstring é um texto dentro da função explicando o que ela faz.

Isso ajuda humanos e também ajuda a documentar ferramentas de agentes.

````python
def count_distinct_values(values):
    """
    Counts the number of distinct values in a list.
    """
    return len(set(values))

products = ["Laptop", "Phone", "Tablet", "Phone", "Laptop"]

print(count_distinct_values(products))
````

## 6.4 Simulando um pequeno registro de tools

Um agente precisa saber quais funções estão disponíveis.

Podemos registrar funções em um dicionário chamado `TOOLS`.

````python
def list_distinct_values(values):
    """
    Lists distinct values.
    """
    return sorted(set(values))

def count_distinct_values(values):
    """
    Counts distinct values.
    """
    return len(set(values))

TOOLS = {
    "list_distinct_values": list_distinct_values,
    "count_distinct_values": count_distinct_values
}

print(TOOLS.keys())
````

````python
# Chamando uma função pelo nome salvo no dicionário TOOLS.

products = ["Laptop", "Phone", "Tablet", "Phone", "Laptop"]

prompt = "count_distinct_values"

tool_function = TOOLS[prompt]

result = tool_function(products)

print(result)
````

# 8. JSON: a ponte entre LLM e Python

Modelos de IA respondem em texto.

Mas um agente precisa transformar esse texto em uma estrutura que Python consiga executar.

JSON é um formato muito usado para isso.

````python
import json

# Isto é uma string em formato JSON.
response = '{"tool": "list_distinct_values", "args": {"column": "product"}}'

print(response)
print(type(response))
````

````python
# json.loads() transforma a string JSON em um dicionário Python.

response = '{"tool": "list_distinct_values", "args": {"column": "product"}}'

decision = json.loads(response)

print(decision)
print(type(decision))
print(decision["tool"])
print(decision["args"])
````

## 8.1 Simulando uma decisão de agente

Vamos simular uma decisão que veio de um modelo.

O Python vai ler a decisão e executar a função correta.

````python
# Criamos algumas funções.

def list_distinct_values(values):
    return sorted(set(values))

def count_distinct_values(values):
    return len(set(values))

# Registramos as funções como tools.
TOOLS = {
    "list_distinct_values": list_distinct_values,
    "count_distinct_values": count_distinct_values
}

# Esta seria a decisão estruturada do modelo.
decision_json = '{"tool": "list_distinct_values", "args": {}}'

# Transformamos JSON em dicionário.
decision = json.loads(decision_json)

# Dados de exemplo.
products = ["Laptop", "Phone", "Tablet", "Phone", "Laptop"]

# Pegamos o nome da tool.
tool_name = decision["tool"]

# Buscamos a função.
tool_function = TOOLS[tool_name]

# Executamos.
result = tool_function(products)

print(result)
````

## 8.2 Tratando JSON inválido

Às vezes, um modelo pode devolver um texto que não é JSON válido.

Um agente robusto precisa lidar com isso.

````python
import json

bad_response = "this is not valid json"

try:
    decision = json.loads(bad_response)
    print(decision)
except json.JSONDecodeError:
    print("Invalid JSON. Falling back to a safe response.")
````

## 10.1 Normalizando nomes de colunas

Vamos criar uma função para mapear termos comuns para nomes reais do dataset.

````python
def normalize_column_name(column):
    """
    Maps natural language column references to real column names.
    """
    if column is None:
        return column

    # Padroniza o texto.
    normalized = column.strip().lower().replace(" ", "_")

    # Mapeia nomes comuns para nomes reais.
    aliases = {
        "sales": "revenue",
        "total_sales": "revenue",
        "sales_amount": "revenue",
        "total_revenue": "revenue",
        "units": "units_sold",
        "units_sold": "units_sold",
        "units_sold_per_product": "units_sold",
        "quantity": "units_sold",
        "product_name": "product",
        "item": "product"
    }

    return aliases.get(normalized, normalized)

print(normalize_column_name("total sales"))
print(normalize_column_name("sales"))
print(normalize_column_name("units sold"))
print(normalize_column_name("item"))
````

## 10.2 Validando depois de normalizar

Agora podemos normalizar primeiro e validar depois.

````python
allowed_columns = ["date", "region", "product", "units_sold", "price", "revenue"]

user_column = "total sales"

column = normalize_column_name(user_column)

if column in allowed_columns:
    print("Column exists:", column)
else:
    print("Column not found:", column)
````

## 10.3 Criando uma função segura para agregar dados

Vamos criar uma função simples que:

1. recebe uma coluna de agrupamento
2. recebe uma coluna numérica
3. recebe uma operação
4. valida tudo
5. executa o cálculo

````python
def simple_groupby(df, group_by, target_column, operation):
    """
    Groups a DataFrame by one column and aggregates a target column.
    This is a simplified version of a data tool used by an agent.
    """
    allowed_operations = ["sum", "mean", "count", "min", "max", "median"]

    # Normaliza nomes de colunas.
    group_by = normalize_column_name(group_by)
    target_column = normalize_column_name(target_column)

    # Valida se as colunas existem.
    if group_by not in df.columns:
        return f"Column not found: {group_by}"

    if target_column not in df.columns:
        return f"Column not found: {target_column}"

    # Valida a operação.
    if operation not in allowed_operations:
        return f"Invalid operation: {operation}"

    # Executa a agregação.
    result = df.groupby(group_by)[target_column].agg(operation).reset_index()

    return result

simple_groupby(df, group_by="product", target_column="total sales", operation="sum")
````

## 10.4 Agrupando por múltiplas colunas

Clientes podem pedir combinações diferentes, por exemplo:

- receita por produto
- receita por região
- receita por produto e região

Para isso, podemos permitir que `group_by` seja uma lista.

````python
def flexible_groupby(df, group_by, target_column, operation):
    """
    Groups a DataFrame by one or more columns and aggregates a target column.
    This is closer to what a real agent tool would need.
    """
    allowed_operations = ["sum", "mean", "count", "min", "max", "median"]

    # Se group_by for texto, transformamos em lista.
    if isinstance(group_by, str):
        group_by = [group_by]

    # Normaliza todos os nomes de agrupamento.
    group_by = [normalize_column_name(col) for col in group_by]

    # Normaliza a coluna alvo.
    target_column = normalize_column_name(target_column)

    # Valida colunas de agrupamento.
    for col in group_by:
        if col not in df.columns:
            return f"Column not found: {col}"

    # Valida coluna alvo.
    if target_column not in df.columns:
        return f"Column not found: {target_column}"

    # Valida operação.
    if operation not in allowed_operations:
        return f"Invalid operation: {operation}"

    # Executa a agregação.
    result = df.groupby(group_by)[target_column].agg(operation).reset_index()

    return result
````

# 11. Juntando as peças: uma mini-simulação de agente

Ainda não vamos construir o agente completo.

Mas já podemos simular o fluxo mental:

1. usuário faz um pedido
2. Python tenta resolver pedidos óbvios
3. caso contrário, uma decisão estruturada seria gerada
4. a função correta é executada
5. o resultado é devolvido

````python
# Tools disponíveis para a mini-simulação.

def list_distinct_products():
    values = sorted(set(df["product"]))
    return "Distinct products:\n" + "\n".join(values)

def count_distinct_products():
    count = df["product"].nunique()
    return f"Number of distinct products: {count}"

TOOLS = {
    "list_distinct_products": list_distinct_products,
    "count_distinct_products": count_distinct_products
}

def deterministic_route(user_input):
    """
    Simple deterministic router.
    It handles obvious requests without using an LLM.
    """
    text = user_input.lower().strip()

    if "distinct products" in text and ("list" in text or "show" in text):
        return {
            "tool": "list_distinct_products",
            "args": {}
        }

    if "distinct products" in text and ("count" in text or "number" in text):
        return {
            "tool": "count_distinct_products",
            "args": {}
        }

    return None

def run_mini_agent(user_input):
    """
    Runs a very small simulation of an agent.
    """
    decision = deterministic_route(user_input)

    if decision is None:
        return "This request would be sent to the LLM in the full agent."

    tool_name = decision["tool"]
    args = decision["args"]

    if tool_name not in TOOLS:
        return f"Tool not found: {tool_name}"

    tool_function = TOOLS[tool_name]

    return tool_function(**args)

print(run_mini_agent("list the distinct products"))
print()
print(run_mini_agent("count the number of distinct products"))
print()
print(run_mini_agent("show me the sales per product"))
````

## 11.1 Mini-simulação com JSON

Agora vamos simular uma decisão em JSON.

Essa é a forma que um LLM poderia responder em um agente real.

````python
import json

decision_json = '''
{
    "tool": "count_distinct_products",
    "args": {}
}
'''

decision = json.loads(decision_json)

tool_name = decision["tool"]
args = decision["args"]

result = TOOLS[tool_name](**args)

print(result)
````
